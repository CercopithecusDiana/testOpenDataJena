```python
data = pd.read_csv('data/corona_erkrankungen_jena.csv')
data_erfurt = pd.read_csv('data/corona_erkrankungen_erfurt.csv')
data_gera =  pd.read_csv('data/corona_erkrankungen_gera.csv')
```

```python
corona_zeit = np.array(data['zeit'])
corona_zeit_erfurt = np.array(data_erfurt['zeit'])
corona_zeit_gera = np.array(data_gera['zeit'])
#print(corona_zeit)
```

```python
corona_erkrankte = np.array(data['erkrankte'])
corona_erkrankte_erfurt = np.array(data_erfurt['erkrankte'])
corona_erkrankte_gera = np.array(data_gera['erkrankte'])
#print(corona_erkrankte)
#print(corona_erkrankte_erfurt)
#print(corona_erkrankte_gera)
```

```python
corona_genesene = np.array(data['genesene'])
corona_genesene_erfurt = np.array(data_erfurt['genesene'])
corona_genesene_gera = np.array(data_gera['genesene'])
#print(corona_genesene)
#print(corona_genesene_erfurt)
#print(corona_genesene_gera)
```

```python
corona_tote = np.array(data['tote'])
corona_tote_erfurt = np.array(data_erfurt['tote'])
corona_tote_gera = np.array(data_gera['tote'])
# print(corona_tote)
```

```python
seaborn.set()
plt.title('Corona Jena')
plt.ylabel('Fälle')
plt.xlabel('Zeit')
plt.plot(corona_zeit, corona_erkrankte)
plt.plot(corona_zeit, corona_genesene)
plt.plot(corona_zeit, corona_tote)
```

```python
seaborn.set()
plt.title('Corona Erfurt')
plt.ylabel('Fälle')
plt.xlabel('Zeit')
plt.plot(corona_zeit_erfurt, corona_erkrankte_erfurt)
plt.plot(corona_zeit_erfurt, corona_genesene_erfurt)
plt.plot(corona_zeit_erfurt, corona_tote_erfurt)
```

```python
seaborn.set()
plt.title('Corona Gera')
plt.ylabel('Fälle')
plt.xlabel('Zeit')
plt.plot(corona_zeit_gera, corona_erkrankte_gera)
plt.plot(corona_zeit_gera, corona_genesene_gera)
plt.plot(corona_zeit_gera, corona_tote_gera)
```
